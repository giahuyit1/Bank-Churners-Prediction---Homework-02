```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
import models as mdl
```

```python
# LOAD PROCESSED DATA
X_path = "../data/processed/X_numpy.csv"
Y_path= "../data/processed/Y_numpy.csv"
X = np.genfromtxt(X_path, delimiter=",", dtype=float, skip_header=0)
Y = np.genfromtxt(Y_path, delimiter=",", dtype=float, skip_header=0)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.25)
```

```python
# model = LogisticRegression()
model = LogisticRegression(class_weight="balanced")

model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
print(Y_pred)
grade = accuracy_score(Y_test,Y_pred) *100
print(grade)
```

```python
np.savetxt('../data/processed/Y_predict_numpy.csv', 
           Y_pred, 
           delimiter=',', # Chỉ định dấu phân cách là dấu phẩy
           fmt='%f')      # Chỉ định định dạng số là float
print("Đã Xuất ra file CSV")
```
